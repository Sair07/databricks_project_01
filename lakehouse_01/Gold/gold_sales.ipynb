# Transformation


```python
%sql
SELECT
    sd.order_number,
    pr.product_key,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM workspace.silver.crm_sales sd
LEFT JOIN gold.dim_products pr
    ON sd.product_number = pr.product_key
LEFT JOIN gold.dim_customers cu
    ON sd.customer_id = cu.customer_id;
```

```python
query= """
SELECT
    sd.order_number,
    pr.product_key,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM workspace.silver.crm_sales sd
LEFT JOIN gold.dim_products pr
    ON sd.product_number = pr.product_key
LEFT JOIN gold.dim_customers cu
    ON sd.customer_id = cu.customer_id;
"""
df = spark.sql(query)
```

# Writing to gold

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_sales")
```
