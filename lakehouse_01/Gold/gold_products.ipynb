#Transformation Logic

Query to join multiple tables

```python
%sql
SELECT
    ROW_NUMBER() OVER (ORDER BY pn.product_start_date, pn.product_key) AS product_number_key, -- Surrogate key
    pn.product_id,
    pn.product_key,
    pn.product_name,
    pn.cat_id,
    pc.category,
    pc.sub_category,
    pc.maintenance,
    pn.product_line,
    pn.product_start_date
FROM workspace.silver.crm_prod pn
LEFT JOIN workspace.silver.erp_cat pc
    ON pn.cat_id = pc.id
--WHERE pn.end_date IS NULL; -- Filter out all historical data
```

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY pn.product_start_date, pn.product_key) AS product_number_key, -- Surrogate key
    pn.product_id,
    pn.product_key,
    pn.product_name,
    pn.cat_id,
    pc.category,
    pc.sub_category,
    pc.maintenance,
    pn.product_line,
    pn.product_start_date
FROM workspace.silver.crm_prod pn
LEFT JOIN workspace.silver.erp_cat pc
"""
df = spark.sql(query)
```

#writing to gold

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
