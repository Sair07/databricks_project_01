# Loading data from volume

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

Loading Cust data from volume

```python
df = (
    spark.read.option("header", "true").option("inferschema","true").csv("/Volumes/workspace/bronze/source_systems/source_crm/cust_info.csv")
)
display(df)
```

Write to bronze

```python
df.write.mode("overwrite").saveAsTable("workspace.bronze.crm_cust_info")
```

Reading all files using a list of dict

```python
INGESTION_CONFIG = [
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_systems/source_crm/cust_info.csv",
        "table": "crm_cust_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_systems/source_crm/prd_info.csv",
        "table": "crm_prd_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_systems/source_crm/sales_details.csv",
        "table": "crm_sales_info"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_systems/source_erp/CUST_AZ12.csv",
        "table": "erp_cust_az12"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_systems/source_erp/LOC_A101.csv",
        "table": "erp_loc_a101"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_systems/source_erp/PX_CAT_G1V2.csv",
        "table": "erp_px_cat"
    }
]
```

```python
for item in INGESTION_CONFIG:
    print(item['path'])
```

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} → workspace.bronze.{item['table']}")

    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"workspace.bronze.{item['table']}")
    )
```
