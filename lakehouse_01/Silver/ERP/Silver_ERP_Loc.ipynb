#Reading Data From Bronze

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.types import DateType
```

```python
df = spark.table('workspace.bronze.erp_loc_a101')
```

```python
display(df)
```

```python
for field in df.schema.fields:
  print(field.name, field.dataType)
```

#Transforming Data

Transforming sales table
- trim string values - good practice
- normalization for status & gender
- column names are not friendly

##Trimming

```python
#Triming the string column 

for field in df.schema.fields:
    if field.dataType == StringType():
        df = df.withColumn(field.name, trim(col(field.name)))
```

## Correcting the gender column

```python
#standardizing the columns 
#lets get all the unique values to understand what all Values we have in gender column 
df.select("CNTRY").distinct().show()
```

```python
#renaming the column names
remap_name = {
    "CID": "customer_key",
    "CNTRY": "country"
}
for old_val, new_val in remap_name.items():
    df = df.withColumnRenamed(old_val, new_val)
#validating
df.columns
```

```python
#removing the - from customer_key

df = df.withColumn("customer_key", regexp_replace(col("customer_key"), "-", "")) 
```

```python
display(df)
```

#Writing to Silver 

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.silver.erp_loc")
```
